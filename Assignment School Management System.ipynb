```python
# Character Creation
def Character_Creation():
    character_id = input('Enter character id: ')
    character_name = input('Enter character name: ')
    character_type = input('Enter character type: ')
    character_class = input('Enter character class: ')
    character_age = input('Enter character age: ')
    character_status = input('Enter character status: ')
    character_kids = input('Enter character kids: ')
    character_years = input('Enter character years: ')
  
    character_data = {}
    character_data['id'] = character_id.title()
    character_data['name'] = character_name.title()
    character_data['type'] = character_type.title()
    character_data['class'] = character_class.title()
    character_data['age'] = character_age.title()
    character_data['status'] = character_status.title()
    character_data['kids'] = character_kids.title()
    character_data['years'] = character_years.title()
    dict_character[character_id]= character_data
    print(dict_character)
```

```python
def Character_Updation():
    update_data = input('Enter character id: ')
    try:
        character_id = update_data
        character_name = input('Enter character name: ')
        character_type = input('Enter character type: ')
        character_class = input('Enter character class: ')
        character_age = input('Enter character age: ')
        character_status = input('Enter character status: ')
        character_kids = input('Enter character kids: ')
        character_years = input('Enter character years: ')
        dict_character[character_id]['id']= character_id.title()
        dict_character[character_id]['name']= character_name.title()
        dict_character[character_id]['type']= character_type.title()
        dict_character[character_id]['class']= character_class.title()
        dict_character[character_id]['age']= character_age.title()
        dict_character[character_id]['status']= character_status.title()
        dict_character[character_id]['kids']= character_kids.title()
        dict_character[character_id]['years']= character_years.title()
    except:
        print('\nRecord not found\n')    
```

```python
def Character_Deletion():
    del_data = input('Enter character id: ')
    try:
        del dict_character[del_data]
        print(dict_character)
    except:
        print('\nRecord not found\n')   
```

```python
def Ask_User():
    i = ''
    while i != '1':
        print('Press 1 to enter a new character record.')
        print('Press 2 to update an existing record.')
        print('Press 3 to search from the character records.')
        print('Press 4 to remove a record.')
        print('Press q to exit the program.')
        print('\n')
        i = input('Press any key: ')
        if i == '1'or i == '2' or i == '3' or i == '4' or i == 'q':
            return i
        else:
            print('Invalid Input\n')
```

```python
def Ask_User_to_continue():
    i = ''
    while i != '1':
        print('\n')
        print('Press 1 to enter a new character')
        print('Press a to exit this feature')
        print('\n')
        i = input('Press any key: ')
        if i == '1' or i == 'a':
            return i
        else:
            print('\nInvalid Input\n')  
```

```python
def Character_Search():
    print('Press 1 to search using the type of characters')
    print('Press 2 to search using an character_id.')
    print('Press 3 to search using the age.')
    print('Press 4 to search using the status.')
    print('Press 5 to search using the class.')
    user = input('Press any key: ')
    try:
        if user  == '1':
            user_search = input('Please search: ')
            for i in dict_character:
                if dict_character[i]['type'] == user_search.title():
                    print(dict_character[i])      
        elif user  == '2':
            user_search = input('Please search: ')
            for i in dict_character:
                if dict_character[i]['id'] == user_search.title():
                    print(dict_character[i])                           
        elif user  == '3':
            user_search = input('Please search: ')
            for i in dict_character:
                if dict_character[i]['age'] == user_search.title():
                    print(dict_character[i])
        elif user  == '4':
            user_search = input('Please search: ')
            for i in dict_character:
                if dict_character[i]['status'] == user_search.title(): 
                    print(dict_character[i])
        elif user  == '5':
            user_search = input('Please search: ')
            for i in dict_character:
                if dict_character[i]['class'] == user_search.title():
                    print(dict_character[i])   
        else:
            print('\nInvalid Inputaa\n')
    except:
        print('\nInvalid Entry\n')
```
